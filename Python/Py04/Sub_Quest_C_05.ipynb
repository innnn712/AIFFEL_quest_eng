```python
# ì¶œë ¥ë¬¼ (ì€í–‰ì´ë¦„, ì˜ˆê¸ˆì£¼, ê³„ì¢Œë²ˆí˜¸, ì”ì•¡)
import random
from datetime import datetime

class Account:
    account_count = 0

    def __init__(self, depositor, balance):
        self.bank_name = "SCì€í–‰"
        self.depositor = depositor
        self.balance = balance
        
        # ê³„ì¢Œë²ˆí˜¸ : 3-2-6 ëœë¤ ìƒì„±
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        self.account_number = f"{part1}-{part2}-{part3}"
        
        # Q7. ì…ê¸ˆ íšŸìˆ˜ ì¹´ìš´íŠ¸ (ì´ì ì§€ê¸‰)
        self.deposit_count = 0 # ì´ˆê¸°í™”

        # Q10. ì…ì¶œê¸ˆ ë‚´ì—­ ê¸°ë¡
        self.deposit_records = []
        self.withdraw_records = []
        
        # í´ë˜ìŠ¤ ë³€ìˆ˜ ì¦ê°€
        Account.account_count += 1

    @classmethod
    def get_account_num(cls):
        return cls.account_count

    def deposit(self, amount):
        if amount < 1:
            print("ì…ê¸ˆì€ ìµœì†Œ 1ì› ì´ìƒë§Œ ê°€ëŠ¥í•©ë‹ˆë‹¤.")
            return

        self.balance += amount
        self.deposit_count += 1

        # ì…ê¸ˆ ê¸°ë¡ì„ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥
        self.deposit_records.append({
            'datetime': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            'amount': amount
         })

        # Q7. ì…ê¸ˆ 5íšŒ ì‹œ 1% ì´ì ì§€ê¸‰
        if self.deposit_count >= 5:
            interest = int(self.balance * 0.01)
            self.balance += interest
            print(f"ì…ê¸ˆ 5íšŒë¥¼ ë‹¬ì„±í•˜ì…¨ìŠµë‹ˆë‹¤. {interest:,}ì›ì˜ ì´ìê°€ ì§€ê¸‰ë˜ì—ˆìŠµë‹ˆë‹¤.")
            
    # Q5. ì¶œê¸ˆ ë§¤ì„œë“œ
    def withdraw(self, amount):
        if amount > self.balance:
            print("ì”ê³ ê°€ ë¶€ì¡±í•©ë‹ˆë‹¤.")
            return

        self.balance -= amount

        self.withdraw_records.append({
            'datetime': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            'amount': amount
            })

    # Q6. ì •ë³´ ì¶œë ¥ ë§¤ì„œë“œ
    def display_info(self):
        print (f"{self.bank_name} , ì˜ˆê¸ˆì£¼: {self.depositor}, ê³„ì¢Œë²ˆí˜¸: {self.account_number}, ì”ê³ : {self.balance:,}ì›")

    # Q10. ì…ê¸ˆ ë‚´ì—­ / AI support
    def deposit_history(self):
        print(f"\n=== {self.depositor}ë‹˜ì˜ ì…ê¸ˆ ë‚´ì—­ ===")
        if not self.deposit_records:
            print("ì…ê¸ˆ ë‚´ì—­ì´ ì—†ìŠµë‹ˆë‹¤.")
            return

        for i, record in enumerate(self.deposit_records, 1):
            print(f"{i}. {record['datetime']} - {record['amount']:,}ì›")
        print(f"ì´ ì…ê¸ˆ íšŸìˆ˜: {len(self.deposit_records)}íšŒ\n")
    # Q10. ì¶œê¸ˆ ë‚´ì—­ / AI support
    def withdraw_history(self):
        print(f"\n=== {self.depositor}ë‹˜ì˜ ì¶œê¸ˆ ë‚´ì—­ ===")
        if not self.withdraw_records:
            print("ì¶œê¸ˆ ë‚´ì—­ì´ ì—†ìŠµë‹ˆë‹¤.")
            return

        for i, record in enumerate(self.withdraw_records, 1):
            print(f"{i}. {record['datetime']} - {record['amount']:,}ì›")
        print(f"ì´ ì¶œê¸ˆ íšŸìˆ˜: {len(self.withdraw_records)}íšŒ\n")

print("=" * 50)
print("ê³„ì¢Œ ìƒì„±")
print("=" * 50)

accounts = []

acc1 = Account("ìŠ¤í…ŒíŒŒë‹ˆ", 500000)
accounts.append(acc1)

acc2 = Account("ì´ì„¸ëŒ", 1500000)
accounts.append(acc2)

acc3 = Account("ë¨¸ìŠ¤í¬", 800000)
accounts.append(acc3)

acc4 = Account("ìŠ¤í‹°ë¸Œ", 2000000)
accounts.append(acc4)

print(f"\nì´ ìƒì„±ëœ ê³„ì¢Œ ìˆ˜: {Account.get_account_num()}ê°œ\n")

# ê° ê³„ì¢Œì— ì…ì¶œê¸ˆ í…ŒìŠ¤íŠ¸
print("=" * 50)
print("ì…ì¶œê¸ˆ í…ŒìŠ¤íŠ¸")
print("=" * 50)

print("\n[ìŠ¤í…ŒíŒŒë‹ˆ ê³„ì¢Œ]")
acc1.deposit(100000)
acc1.deposit(50000)
acc1.deposit(30000)
acc1.deposit(20000)
acc1.deposit(10000)  # 5íšŒ ì…ê¸ˆ â†’ ì´ì ì§€ê¸‰
acc1.withdraw(50000)

print("\n[ì´ì„¸ëŒ ê³„ì¢Œ]")
acc2.deposit(200000)
acc2.deposit(100000)
acc2.withdraw(300000)

print("\n[ë¨¸ìŠ¤í¬ ê³„ì¢Œ]")
acc3.deposit(150000)
acc3.deposit(50000)
acc3.deposit(100000)
acc3.deposit(200000)
acc3.deposit(50000)  # 5íšŒ ì…ê¸ˆ â†’ ì´ì ì§€ê¸‰

print("\n[ìŠ¤í‹°ë¸Œ ê³„ì¢Œ]")
acc4.withdraw(500000)
acc4.deposit(100000)

# Q9. ì”ê³ ê°€ 100ë§Œì› ì´ìƒì¸ ê³ ê° ì •ë³´ ì¶œë ¥
print("\n" + "=" * 50)
print("ì”ê³ ê°€ 100ë§Œì› ì´ìƒì¸ ê³ ê° ì •ë³´")
print("=" * 50)

for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

# Q10. ì…ì¶œê¸ˆ ë‚´ì—­ ì¶œë ¥
print("\n" + "=" * 50)
print("ê±°ë˜ ë‚´ì—­ ì¡°íšŒ")
print("=" * 50)

acc1.deposit_history()
acc1.withdraw_history()

acc3.deposit_history()
acc3.withdraw_history()
```

# Quest 2. ìë™ì‚¬ëƒ¥ RPG ê²Œì„

```python
import random
from datetime import datetime


# Q1. Character í´ë˜ìŠ¤
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.max_hp = hp
        self.current_hp = hp
        self.attack = attack
        self.defense = defense
    
    # í˜„ì¬ ì²´ë ¥ì´ 0 ì´ìƒì¸ì§€ bool ê°’ì„ ë°˜í™˜í•˜ëŠ” is_alive ë©”ì„œë“œ
    def is_alive(self):
        return self.current_hp > 0
    
    def take_damage(self, damage):
        actual_damage = damage - self.defense
        if actual_damage > 0:
            self.current_hp -= actual_damage
            print(f"{self.name}ì´(ê°€) {actual_damage}ì˜ ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ìŠµë‹ˆë‹¤! (ë‚¨ì€ ì²´ë ¥: {max(0, self.current_hp)})")
        else:
            print(f"{self.name}ì˜ ë°©ì–´ë ¥ìœ¼ë¡œ ê³µê²©ì„ ë§‰ì•˜ìŠµë‹ˆë‹¤!")
    
    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"{self.name}ì´(ê°€) {target.name}ì„(ë¥¼) ê³µê²©! ({damage} ë°ë¯¸ì§€)")
        target.take_damage(damage)


# Q2. Player í´ë˜ìŠ¤
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0
    # ê²½í—˜ì¹˜ íšë“
    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}ì´(ê°€) {exp} ê²½í—˜ì¹˜ë¥¼ íšë“í–ˆìŠµë‹ˆë‹¤! (í˜„ì¬ ê²½í—˜ì¹˜: {self.experience})")
    # ë ˆë²¨ì—… 
    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50
            print(f"\nğŸ‰ ë ˆë²¨ì—…! {self.name}ì˜ ë ˆë²¨ì´ {self.level}ì´ ë˜ì—ˆìŠµë‹ˆë‹¤!")
            print(f"   ê³µê²©ë ¥: {self.attack}, ë°©ì–´ë ¥: {self.defense}\n")


# Q2. Monster í´ë˜ìŠ¤
class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)
        print(f"\n{'='*50}")
        print(f"ğŸ¦¹ {name} ë“±ì¥! (ë ˆë²¨ {level})")
        print(f"   ì²´ë ¥: {hp}, ê³µê²©ë ¥: {attack}, ë°©ì–´ë ¥: {defense}")
        print(f"{'='*50}\n")


# Q3. battle í•¨ìˆ˜
def battle(player, monster):
    print(f"\nâš”ï¸  ì „íˆ¬ ì‹œì‘: {player.name} VS {monster.name}\n")
    
    turn = 1
    while player.is_alive() and monster.is_alive():
        print(f"--- Turn {turn} ---")
        
        # í”Œë ˆì´ì–´ ê³µê²©
        player.attack_target(monster)
        
        # ëª¬ìŠ¤í„°ê°€ ì‚´ì•„ìˆìœ¼ë©´ ë°˜ê²©
        if monster.is_alive():
            monster.attack_target(player)
        
        print()
        turn += 1
    
    # ì „íˆ¬ ê²°ê³¼
    if player.is_alive():
        exp_gained = monster.level * 20
        print(f"âœ¨ ì „íˆ¬ ìŠ¹ë¦¬!")
        player.gain_experience(exp_gained)
        player.level_up()
        
        # ì „íˆ¬ í›„ ì²´ë ¥ íšŒë³µ
        player.current_hp = player.max_hp
        print(f"{player.name}ì˜ ì²´ë ¥ì´ íšŒë³µë˜ì—ˆìŠµë‹ˆë‹¤! (ì²´ë ¥: {player.current_hp})\n")
        return True
    else:
        print(f"ğŸ’€ ì „íˆ¬ íŒ¨ë°°..\n")
        return False


# Q4. main í•¨ìˆ˜
def main():
    # ëª¬ìŠ¤í„° ë”•ì…”ë„ˆë¦¬
    monster_dict = {'ìŠ¬ë¼ì„': 1, 'ê³ ë¸”ë¦°': 2, 'ì˜¤í¬': 3}
    
    # í”Œë ˆì´ì–´ ìƒì„±
    print("="*50)
    print("ğŸ® ìë™ì‚¬ëƒ¥ RPG ê²Œì„ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!")
    print("="*50)
    player_name = input("\ní”Œë ˆì´ì–´ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    player = Player(player_name)
    
    print(f"\n{player.name} ìºë¦­í„°ê°€ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
    print(f"ë ˆë²¨: {player.level}, ì²´ë ¥: {player.current_hp}, ê³µê²©ë ¥: {player.attack}, ë°©ì–´ë ¥: {player.defense}\n")
    
    # ëª¬ìŠ¤í„°ì™€ ì „íˆ¬
    for monster_name, monster_level in monster_dict.items():
        if not player.is_alive():
            break
        
        monster = Monster(monster_name, monster_level)
        
        if not battle(player, monster):
            print("="*50)
            print("ğŸ® ê²Œì„ì˜¤ë²„")
            print("="*50)
            return
    
    # ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ë¬¼ë¦¬ì¹œ ê²½ìš°
    if player.is_alive():
        print("="*50)
        print("ğŸ† ì¶•í•˜í•©ë‹ˆë‹¤! ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ë¬¼ë¦¬ì³¤ìŠµë‹ˆë‹¤!")
        print(f"ìµœì¢… ë ˆë²¨: {player.level}")
        print("="*50)


# ê²Œì„ ì‹¤í–‰
if __name__ == "__main__":
    main()
```
